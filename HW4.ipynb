## Задание 1

Найти область определения функции

$$
z=\sqrt{1-x^3}+\ln{(y^2-1)}
$$



### Решение

Подкоренное выражение неотрицательно. Под знаком логарифма - строго положительно.

$$
\begin{cases}

y^2 - 1 > 0
\end{cases}
$$

$$
\begin{cases}
x \in (-\infty; 1] \\
y \in  (-\infty; -1)\cup(1; \infty)
\end{cases}
$$

## Задание 3

Найти полный дифференциал функции в точке (1;1)

$$
z=\sqrt{2xy + \cos{\frac{x}{y}}}
$$

### Решение

Продиффиренцируем в общем виде.
$$
\begin{cases}
\frac{\partial z}{\partial x} = \frac{1}{2\sqrt{2xy + \cos{\frac{x}{y}}}}(2y + \frac{1}{y}\sin{\frac{x}{y}}) \\
\frac{\partial z}{\partial y} = \frac{1}{2\sqrt{2xy + \cos{\frac{x}{y}}}}(2x - \frac{x}{y^2}\sin{\frac{x}{y}})
\end{cases}
$$

Подставим координаты точки.

$$
dz(1, 1) = \frac{2 + \sin{1}}{2\sqrt{2 + \cos{1}}} + \frac{2 - \sin{1}}{2\sqrt{2 + \cos{1}}}=\frac{2}{\sqrt{2 + \cos{1}}}
$$

```python
import numpy as np
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
```

```python


def makeData ():
    x = np.arange (-1, 3, 0.05)
    y = np.arange (2, 6, 0.05)
    xgrid, ygrid = np.meshgrid(x, y)

    zgrid = xgrid**2 + (xgrid * ygrid) + ygrid**2 - (6 * xgrid) - (9 * ygrid)
    return xgrid, ygrid, zgrid

x, y, z = makeData()

fig = pylab.figure()
axes = Axes3D(fig)

# axes.plot_surface(x, y, z)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap = cm.jet)

pylab.show()
```
