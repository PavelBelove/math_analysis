## Задание 1

Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

### Решение:

Периодические функции не имеют предела на бесконечностях. 

$$f(u) = tg(u)$$
Чтобы убрать предел в нуле, под знаком тангенса должна быть функция, обращающая в бесконечности при приближении к нулю, и в бесконечность.
$$u(x) = \frac{x+1}{x}$$

Таким образом,

$$f(x) = tg(\frac{x+1}{x})$$

Не имеет пределов ни в нуле, ни в бесконечности.

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format='retina'

sns.set(rc={'figure.figsize': (9, 6)})
```

```python

np.seterr(divide='ignore')

x = np.arange(-50, 50, 0.5)
y = np.tan(x + 1/x)

plt.axhline(linewidth=2.0, color="black")
plt.axvline(linewidth=2.0, color="black")

plt.plot(x,y)
plt.show()
```

## Задание 2

Привести пример функции, не имеющей предела в точке, но определенной в ней.

### Решение:
Серьезно задумался, можно ли сказать, что функция из предыдущего примера отвечает этому требованию.
Тангенс определен не на всей области, и если $ \frac{x^2+1}{x} $ попадает в $ \pi/2 +  \pi n$ - функция не определена. При приближении аргумента к нулю, подобные "совпадения" будут все чаще - можно ли тут говорить о том, что функция определена в нуле, или не определена?
Чтобы привести гарантированно правильный ответ, заменю тангенс на синус.

$$f(x) = sin(\frac{x^2+1}{x})$$


```python

np.seterr(divide='ignore')

x = np.arange(-50, 50, 0.5)
y = np.sin(x + 1/x)

plt.axhline(linewidth=2.0, color="black")
plt.axvline(linewidth=2.0, color="black")

plt.plot(x,y)
plt.show()
```

## Задание 3

Исследовать функцию $f(x)=x^3 - x^2$ по плану:
- Область задания и область значений.
- Нули функции и их кратность.
- Отрезки знакопостоянства.
- Интервалы монотонности.
- Четность функции.
- Ограниченность.
- Периодичность.





### Решение

Область задания : $(-\infty; \infty)$
Jбласть значений: $(-\infty; \infty)$ 
Нули функции: $f(x)=x^3 - x^2\implies f(x)=x^2(x - 1)\implies x_1 = 1, x_2 = 0$

Кратность нулей функции: $f'(x)= 3x^2 - 2x\implies f'(0)=0, f'(1)=1$. Функция имеет по одному корню, нули - простые.

Отрезки знакопостоянства: $(-\infty; 0), (0; 1);  (1; \infty)$

Четность функции: Не является ни четной, ни нечетной.

Не ограничена и не периодична.

```python

np.seterr(divide='ignore')

x = np.arange(-1, 2, 0.1)
y = x**3 - x**2

plt.axhline(linewidth=2.0, color="black")
plt.axvline(linewidth=2.0, color="black")
plt.axis('equal')

plt.plot(x,y)
plt.show()
```
